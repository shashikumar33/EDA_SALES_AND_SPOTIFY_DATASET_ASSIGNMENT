```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
df=pd.read_csv(r"C:\Users\shash\Downloads\amazon.csv")
```

```python
df
```

```python
df.info()
```

```python
#first convert some columns that store a int value but it show object types
```

```python
df.rating=pd.to_numeric(df.rating, errors="coerce")
```

```python
df.rating_count=df.rating_count.str.replace(",","")
```

```python
df.rating_count=df.rating_count.astype(float,errors="ignore")
```

```python
df.discounted_price=df.discounted_price.str.replace("₹","")
df['discounted_price'] = df['discounted_price'].str.replace(",","")
df.discounted_price=pd.to_numeric(df.discounted_price)
```

```python
df.actual_price=df.actual_price.str.replace("₹","")
df.actual_price = df.actual_price.str.replace(",","")
df.actual_price=pd.to_numeric(df.actual_price)
```

```python
df.discount_percentage=df.discount_percentage.str.replace("%","")
df.discount_percentage=df.discount_percentage.astype(float,errors="ignore")
```

```python
df.info()
```

```python
df
```

```python
"Q1.  What is the average rating for each product category?"
```

```python
"Q3 What is the distribution of discounted prices vs. actual prices?"
```

```python
# Set up the figure and axis
plt.figure(figsize=(10, 6))

# Plot histogram for actual prices
sns.histplot(df['actual_price'], bins=30, kde=True, color='blue', label='Actual Price')

# Plot histogram for discounted prices on the same axis
sns.histplot(df['discounted_price'], bins=30, kde=True, color='orange', label='Discounted Price')

# Add labels and legend
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()

# Show plot
plt.show()
```

```python
"""1. Identify the top 5 popular artists based on the mean popularity of their tracks.Show the mean popularity of 
tracks for the top 5 popular artists Using BarPlot"""
```

```python
popular_artist=df2.groupby(["Artist","Track ID"])["Popularity"].mean()
```

```python
top_5_popular_artists=popular_artist.sort_values(ascending=False).head(5).reset_index()
```

```python
top_5_popular_artists
```

```python
plt.figure()
sns.barplot(x="Artist",y="Popularity", data=top_5_popular_artists)
plt.title("TOP 5 ARTIST BASED ON THEIR POPULARITY")
plt.xlabel("Artist")
plt.ylabel("popularity")
plt.xticks(rotation=90)
plt.show()
```

```python
""" 2. Determine the top 5 popular songs based on their popularity ratings.Display the popularity ratings of the top 
5 popular songs using BarPlot"""
```

```python
popular_song=df2.groupby("Track Name")["Popularity"].mean()
top_5_popular_songs=popular_song.sort_values(ascending=False).reset_index().head(5)
top_5_popular_songs
```

```python
plt.figure()
sns.barplot(x="Track Name",y="Popularity",data=top_5_popular_songs)
plt.title("top 5 popular song")
plt.xlabel("Song Name")
plt.ylabel("Popularity")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
""" 3. Find the top 5 trending genres based on the mean popularity of tracks within each genre.Visualize the mean 
popularity of tracks for the top 5 trending genres"""
```

```python
top_5_trending_genres=df2.groupby("Artist")["Popularity"].mean()
top_5_trending_genres_list=top_5_trending_genres.sort_values(ascending=False).reset_index().head(5)
top_5_trending_genres_list
```

```python
plt.figure()
sns.barplot(x="Artist", y="Popularity",data=top_5_trending_genres_list)
plt.title('Mean Popularity of Tracks for Top 5 Trending Genres (Artists)')
plt.xlabel('Artist')
plt.ylabel('Mean Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
""" 4. Identify the top 5 longest songs among the tracks of the top 5 popular artists.Represent the duration of the 
top 5 longest songs among the tracks of the top 5 popular artists using BarPlot"""
```

```python
longest_song=df2.groupby(["Track Name","Duration (ms)"])["Popularity"].mean().reset_index()
longest_song=longest_song.sort_values(by="Popularity",ascending=False).reset_index(drop=True).head(5)
longest_song
```

```python
plt.figure()
sns.barplot(x="Track Name",y="Duration (ms)",data=longest_song)
plt.xlabel("Track Name")
plt.ylabel("Duration")
plt.title("the top 5 longest songs among the tracks of the top 5 popular artists")
plt.xticks(rotation=45)
plt.show()
```
